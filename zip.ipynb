```python
import os
import zipfile

def zip_folder(folder_path, zip_name):
    """
    Zips the entire folder including subfolders.
    
    folder_path : Path of folder to zip
    zip_name : Name of output zip file
    """

    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                
                file_path = os.path.join(root, file)

                # keeps folder structure inside zip
                arcname = os.path.relpath(file_path, folder_path)

                zipf.write(file_path, arcname)

    print(f"Folder '{folder_path}' successfully zipped as '{zip_name}'")


# Example usage
folder_to_zip = "D:\super-resolution\logs"
output_zip = "logs.zip"

zip_folder(folder_to_zip, output_zip)
```
